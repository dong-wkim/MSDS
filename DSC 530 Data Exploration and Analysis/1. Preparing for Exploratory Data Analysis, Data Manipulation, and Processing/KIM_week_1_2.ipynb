# Exercise 6

The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called daily number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot that contains data from January 1, 2020 through September 18, 2020. Clean and pivot the data so that it is in wide format:



- [x] Read in the covid19_cases.csv file.
- [x] Create a date column using the data in the dateRep column and the pd.to_datetime() function.
- [x] Set the date column as the index and sort the index.
- [x] Replace all occurrences of United_States_of_America and United_Kingdom with USA and UK, respectively. Hint: the replace() method can be run on the dataframe as a whole.
- [x] Using the *countriesAndTerritories* column, filter the cleaned COVID-19 cases data down to Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA.
- [x] Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the case counts (the cases column). Be sure to fill in NaN values with 0.

```python
# 1. Prepare and read the data in the covid19_cases.csv file
df = pd.read_csv('data/covid19.csv')
df = pd.DataFrame(df)

# 2. Create a date column

df['dateRep'] = pd.to_datetime(df['dateRep'], format = '%d/%m/%Y')

# 3. Set the date column as the index and sort the index
df = df.set_index('dateRep').sort_index()

# 4. Use the replace() method to update all occurrences of United_States_of_America and United_Kingdom to USA and UK, respectively.

df.replace('United_States_of_America','USA')
df.replace('United_Kingdom','UK')

# 5. Using the countriesAndTerritories column, filter the cleaned COVID-19 cases data down to Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA.

countries = ['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'USA', 'UK']
df = df[df['countriesAndTerritories'].isin(countries)]
df['countriesAndTerritories'].nunique()

# 6. Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the case counts (the cases column). Be sure to fill in NaN values with 0.

df = pd.pivot_table(
    df,
    index = 'dateRep',
    columns = 'countriesAndTerritories', 
    values = 'cases',
    aggfunc = 'sum',
    fill_value = 0
)

df.head()
```
